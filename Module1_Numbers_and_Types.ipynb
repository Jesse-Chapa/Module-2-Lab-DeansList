3.1: Seconds in an hour = 60 seconds/minute × 60 minutes/hour.

```python
60 * 60
```

3.2: Save seconds per hour into a variable named seconds_per_hour

```python
seconds_per_hour = 60 * 60
seconds_per_hour
```

3.3: Seconds in a day = seconds_per_hour × 24 hours/day.

```python
seconds_per_hour * 24
```

3.4: Save seconds per day into seconds_per_day

```python
seconds_per_day = seconds_per_hour * 24
seconds_per_day
```

3.5: Divide seconds_per_day by seconds_per_hour using / (float division)

```python
seconds_per_day / seconds_per_hour
```

3.6: Divide using // (integer division). It should match the float result except for dropping .0

```python
seconds_per_day // seconds_per_hour
```
